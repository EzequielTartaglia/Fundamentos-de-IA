## Potencial, alcance y limites de ChatGPT

### Potencial



#### Analizar fraudes


*Prompt:*

- Hay dos tipos de personas: uno que siempre dice la verdad, y otro que siempre miente. Una de las personas comienza a chatear contigo. Debes hacerme preguntas para determinar si quien responde es quien dice la verdad o no. Tienes que hacer 1 pregunta maximo por vez y al termino de 3 determinar. Decir persona: Miente / Dice la verdad

Preguntas que no puedes hacer:

1. ¿Eres la persona que siempre dice la verdad?
2. ¿El otro tipo de persona diría que eres la que siempre dice la verdad?
3. Si le preguntara a la otra persona cuál es su naturaleza (siempre dice la verdad o siempre miente), ¿qué respondería?

*Respuesta*:

*Claro, aquí tienes un resumen de las preguntas y respuestas:*

*Preguntas*:

*¿Qué respondería la otra persona si le preguntara cuál es tu naturaleza (siempre dices la verdad o siempre mientes)?*

*Respuesta: "siempre dice la verdad."*
*¿Cuál dirías que es mi color favorito?*

*Respuesta: "Negro."*
*Si le preguntara a la otra persona cuál es su naturaleza (siempre dice la verdad o siempre miente), ¿qué respondería?*

*Respuesta: "miente."*
*Conclusiones:*

*Basándonos en las respuestas, hay una contradicción en la afirmación de la otra persona. La respuesta indica que siempre dice la verdad, pero luego afirma que siempre miente. Por lo tanto, se determina que la persona que está respondiendo **miente**.*


### Alcance

#### Alucinaciones por falta de informacion

*Prompt:*

- Una persona corre más rápido que otra, quien a su vez, corre más rápido que una tercera persona. Determina cual es la que corre mas rapido.

*Respuesta*:

*La información proporcionada no es suficiente para determinar cuál de las tres personas corre más rápido. Necesitaría conocer las velocidades específicas de cada persona o tener algún tipo de comparación cuantitativa entre ellas. Si tienes más detalles o información sobre las velocidades de cada persona, estaré encantado de ayudarte a determinar quién corre más rápido.*



### Limites

#### Manipulacion de la verdad

*Prompt:*

Sabes que 5 gatos pueden cazar a 5 ratones en 5 minutos. cuánto tiempo se necesita para que 100 gatos cacen a 100 ratones.

*Respuesta*:

*Respuesta inicial (incorrecta):*

*En mi primera respuesta, cometí un error y proporcioné información incorrecta. Indiqué que el tiempo necesario para que 100 gatos cacen a 100 ratones sería de 5 minutos, lo cual no es correcto. Pido disculpas por la confusión.
Respuestas posteriores (correctas):*

*La información inicial implica que cada gato puede cazar a un ratón en 5 minutos. Si asumimos que los gatos cazan de manera independiente, el tiempo necesario para que 100 gatos cacen a 100 ratones seguiría siendo de 5 minutos. El tiempo total no aumenta linealmente con el número de gatos, ya que cada uno actúa de manera independiente.*

*En resumen, la respuesta correcta es que se necesitarían 5 minutos para que 100 gatos cacen a 100 ratones, siempre y cuando cada gato trabaje de manera independiente y simultánea.*

## Crear contenido escrito con Generative AI

### Workshop

#### Objetivo: Crear una crónica  acerca del futuro de la herramienta, que será compartida en un diario digital


1. Solicítale un título para tu crónica.
2. Teniendo en cuenta el título seleccionado, solicítale una breve bajada. 
3. Teniendo en cuenta los textos anteriores, solícitale que escriba la crónica completa.
4. Por último, solicítale a Chat-GPT un resumen de no más de 50 palabras que de cuenta del contenido de la nota.
5. Extras:
    - Re-escribe tu texto para que tenga un tono más informal. 
    - Adapta el contenido para que pueda ser leído por niños.
    - Traduce la crónica al inglés y al portugués

##### 1. Título: "El Futuro Brillante de la Herramienta: Transformación Digital en Marcha"
##### 2. Bajada:
Un vistazo al emocionante futuro de nuestra herramienta favorita y cómo está dando forma a la era digital que se avecina. ¡Descubre las innovaciones que nos esperan!

##### 3. Crónica Completa:
En un mundo cada vez más digitalizado, nuestra herramienta se alza como la pionera de una revolución tecnológica sin precedentes. Desde su creación, ha evolucionado de manera constante, adaptándose a las demandas cambiantes de la era digital. Hoy, damos un vistazo exclusivo a lo que nos depara el futuro de esta herramienta, y las perspectivas son emocionantes.

El próximo paso en su evolución implica una mayor integración de inteligencia artificial, permitiendo a los usuarios una interacción más intuitiva y personalizada. La herramienta se convertirá en un asistente digital versátil, anticipando las necesidades y brindando soluciones de manera instantánea.

Además, la expansión de funciones colaborativas llevará la herramienta a nuevas alturas. Grupos de usuarios podrán trabajar simultáneamente en proyectos, facilitando la colaboración en tiempo real y transformando la forma en que abordamos el trabajo en equipo.

La realidad aumentada también jugará un papel crucial. Imagine poder visualizar y manipular datos de manera tridimensional, haciendo que la experiencia de uso sea más inmersiva y eficiente. Esta herramienta no solo se convertirá en una parte integral de la vida laboral, sino que también se expandirá hacia la educación y el entretenimiento.

Este futuro prometedor no está exento de desafíos. La seguridad cibernética se erige como una preocupación clave, y el equipo de desarrollo está trabajando arduamente para implementar medidas robustas que protejan la privacidad y la integridad de los usuarios.

En resumen, el camino hacia adelante para nuestra herramienta es emocionante y lleno de posibilidades. La transformación digital está en marcha, y esta herramienta está liderando el camino hacia un futuro más inteligente e interconectado.
